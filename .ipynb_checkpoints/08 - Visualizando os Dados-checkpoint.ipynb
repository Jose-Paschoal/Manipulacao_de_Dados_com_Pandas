# Plotagem e Visualização

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('tested.csv', sep=',')
```

```python
df
```

```python
# criando um gráfico - personalização

plt.rcParams['figure.figsize'] = (11,7)
```

## Dados em Linhas

```python
plt.plot(df.Age, '*-')
plt.xlabel('Passageiro')
plt.show()
```

```python
# alterações no plot - mudança no tamanho da legenda

plt.plot(df.Age, '*-')
plt.xlabel('Passageiro', size=14)
plt.ylabel('Idade', size=16)
plt.show()
```

```python
# 'esticando' o gráfico base

plt.rcParams['figure.figsize'] = (20,7)
```

```python
plt.plot(df.Age, '*-')
plt.xlabel('Passageiro', size=14)
plt.ylabel('Idade', size=16)
plt.show()
```

```python
# outra forma de visualização

df.Age.plot()
plt.show()
```

## Scatter Plot

```python
plt.scatter(df.PassengerId, df.Age)
plt.show()
```

```python
plt.scatter(df.PassengerId, df.Age, color='r')
plt.show()
```

```python
# mudando os marcadores

plt.scatter(df.PassengerId, df.Age, color='r', marker='>')
plt.show()
```

## Histogramas

```python
df.Age.describe()
```

```python
# visual da distribuição das idades

df.Age.hist()
plt.show()
```

## Salvando uma Figura

```python
df.Age.hist()
plt.savefig('distribuicao_idade.png')
plt.show()
```
